```r
Searchword <- "#Influenza OR @Influenza OR #flu OR @flu"  #Tried different Searchwords and this is the best Searchword for which we got collected number of tweets

Tweets <- searchTwitter(Searchword, n=20000, geocode='39.828175,-98.5795,1600mi')
```

```r
# Convert search result tweets into dataframe

Tweets_DF <- twListToDF(Tweets)
```

```r
head(Tweets_DF)
```

```r
# Lookup screen names from this dataframe

Twitter_User_Name <- Tweets_DF$screenName
head(Twitter_User_Name)
```

```r
# Convert these screen names into another dataframe

Users <- lookupUsers(Twitter_User_Name) 
Users_DF <- twListToDF(Users)
```

```r
head(Users_DF)
```

```r
# Removing blank spaces and NA. Collecting tweets only with location

Map_Coordinates <- Users_DF[!is.na(Users_DF$location),]
Map_Coordinates <- Map_Coordinates[!Map_Coordinates$location=="",]
head(Map_Coordinates$location)
```
