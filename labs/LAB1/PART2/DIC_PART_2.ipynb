## PLOT - 1

```r
library("ggplot2")
library("reshape2")

table_df <- read.csv("C:/Users/harik/Desktop/DIC/2/one.csv",header =T)

melt_df <- melt(table_df,id=c("Week"))

head(melt_df)

```

```r
options(repr.plot.width=15, repr.plot.height=10) # Plot dimensions
```

```r
ggplot()+
  geom_bar(data=subset(melt_df,variable %in% c("Total.A","Total.B")),aes(x=factor(Week),y=value,fill=variable),stat="identity")+
  scale_fill_manual(values = c("Total.A" = "#EAF22B", "Total.B" = "#16AB15"),labels=c("A","B")) +
  
  geom_line(data=subset(melt_df, variable %in% c("Percent.Positive.A","Percent.Positive.B","X..Positive")),
            aes(x=factor(Week),y=value*600,group=variable,colour=variable,linetype=variable))+
  
  scale_colour_manual(values=c("Percent.Positive.A"="#E96D12","Percent.Positive.B"="#115502","X..Positive"="#000000"))+
  scale_linetype_manual(values=c("Percent.Positive.A"="dashed","Percent.Positive.B"="dotted","X..Positive"="solid")) +

  scale_y_continuous(sec.axis = sec_axis(~.*0.0014, name = "Percent Positive",breaks = round(seq(2, 30, by = 2),1)))+
  
  labs(x="Week",y="Number of Positive Specimens")+
  
  ggtitle("Influenza Positive Tests Reported to CDC by U.S. Clinical Laboratories, 
          National Summary, 2017-2018 Season")+theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

```r
library("ggmap")
library("maptools")
library("maps")
library("ggplot2")

rm(list=ls())

data_location <- read.csv(file="D:/Spring 2018/DIC/labs/lab1/finalsubmission/DICLAB1/PART2/7.csv")

head(data_location)
```

```r
data_location$levels = sapply(data_location$ACTIVITY.LEVEL, function(x) gsub("Level ", "", as.character(x)))
```

```r
data_location$levels = as.numeric(data_location$levels)
head(data_location,100)
```

```r
state = c()
state_level = c()
for(i in unique(data_location$STATENAME)){
    tempdataframe = data_location[which(data_location$STATENAME == i),]
    state = append(state, i)
    state_level = append(state_level, max(tempdataframe$levels))
}

state_and_levels = data.frame(state, state_level)
```

```r
referencedata = map_data("state")
head(referencedata)
```

```r
region_level = c()
for(i in 1:nrow(referencedata)){
    region = referencedata$region[i]
    index = grep(region, tolower(state_and_levels$state))
    region_level = append(region_level, state_and_levels$state_level[index[1]])
}
```

```r
referencedata$region_level = region_level
```

```r
table(referencedata$region_level)
```

```r
referencedata$final_regionlevel = sapply(referencedata$region_level, function(x) ifelse(x %in% c(0,2), "Minimal", ifelse(x %in% c(4,6), "Low", ifelse(x %in% c(7,8), "Moderate", "High"))))
```

```r

ggplot(referencedata, aes(long, lat)) +
  geom_polygon(aes(group = group, fill = final_regionlevel)) + 
  scale_fill_manual(values = c("#ff471a","#ffff99","#00ff00","#ffa64d"))+coord_quickmap()
```
